Applying one-hot encoding on the genres so that we can get two matrix forms: genres (Columns) to userids(rows)(for creating user profiles) and genres (Columns) to bookIds(rows)(for creating item profiles)

```python
one_hot = pd.get_dummies(df_req['genre'])
```

```python
df_req = df_req.drop('genre',axis = 1)
```

```python
df_req = df_req.join(one_hot)
```

```python
df_req.head()
```

```python
df_req[df_req.user_id=='568ce0dddc00192a814476ab5441acbb']
```

```python
#Adding some more columns genres with respect to the genres from the original datasets from goodreads profile
df_req['Graphics'] = df_req['comics']
df_req['Adventure'] = df_req['comics']
df_req['Fantasy'] = df_req['Horror']
df_req['Mystery'] = df_req['Thriller']
df_req['Crime'] = df_req['Thriller']
```

```python
df_req.head ()
```

```python
#Now we can see that the same userid - bookid- rating combo is appeared 4 times and we can merge them to form a feature vector
df_req.shape
```

```python
grouped_df = df_req.groupby(['user_id','book_id','rating'],as_index=False).sum()
```

# User Feature Vector

Creating user profiles: In a content based approach like ours, user profiles are created to find out the User interestes according to their ratings per genres. This is done by doing a dot product of the user ratings with the genres and normalizing each genre value with sum of ratings given by the user for a book. This process is hard coded below.

Exploring the dataframe a little more

```python
#fetching all the ratings given by the user for all the books read by him
grouped_df [grouped_df.user_id == '396fe0f07a53ad6be40e232c28a35e18']
```

```python
grouped_df.shape
```

```python
#Exploring the dataset if the ratings value is given as '0'. these will not be useful in the computations
#and it simply means that user has not rated any books that they read. These rows will be removed in future as we cannot build user profiles 
#for these users as we are unsure of their interests and if they really liked the book or not
grouped_df[grouped_df['rating']==0].shape
```

```python
grouped_df [grouped_df.user_id == 'fffcf6da0f39d7ab624e2a8da054d2c3']
```

```python
grouped_df.describe()
```

```python
grouped_df.shape
```

```python
# multiply rating with each row value of all the genres

# grouped_df[["Horror","Romance","Thriller","comics"]].multiply (grouped_df["rating"],axis="index")
#for idx, rows in grouped_df.iterrows():
#    for _i in ['Horror','Romance','Thriller','comics']:
#        rows[_i] = rows[_i] * rows['rating']
multiplied_ratings_df  = grouped_df[["Horror","Romance","Thriller","comics","Graphics","Adventure","Fantasy","Mystery","Crime"]].multiply (grouped_df["rating"],axis="index")
new_grouped_df = grouped_df.copy()
new_grouped_df [["Horror","Romance","Thriller","comics","Graphics","Adventure","Fantasy","Mystery","Crime"]] = grouped_df[["Horror","Romance","Thriller","comics","Graphics","Adventure","Fantasy","Mystery","Crime"]].multiply (grouped_df["rating"],axis="index")
```

Here I am doing dot product of user ratings with the one hot encoded genre values

example:
User ratings = 7 for movie which has vector [ 0 0 1 1 0 1] becomes [0 0 7 7 0 7] after dot product

```python
new_grouped_df [new_grouped_df.user_id == '396fe0f07a53ad6be40e232c28a35e18']
```

```python
new_grouped_df [new_grouped_df.user_id == '0001085188e302fc6b2568de45a5f56b']
```

```python
new_grouped_df.shape
```

```python
# for each user summation of ratings per genre and get feature vector 
each_user_sum_of_ratings_df_with_book_id_and_rating = new_grouped_df.groupby(['user_id'],as_index=False).sum()
```

```python
each_user_sum_of_ratings_df = each_user_sum_of_ratings_df_with_book_id_and_rating.drop (['book_id','rating'],axis=1)
```

```python
each_user_sum_of_ratings_df.head ()
```

```python
each_user_sum_of_ratings_df.shape
```

```python
#Adding the sum column which can be used for normalization
each_user_sum_of_ratings_df['Sum'] = each_user_sum_of_ratings_df[["Horror","Romance","Thriller","comics","Graphics","Adventure","Fantasy","Mystery","Crime"]].sum(axis=1)
#each_user_sum_of_ratings_df.Sum.loc[(each_user_sum_of_ratings_df['Sum'] == 0)] = 1
each_user_sum_of_ratings_df.head ()
```

```python
#Fetching columns that has sum as zero due to the fact that some users have not rated the books.
each_user_sum_of_ratings_df[each_user_sum_of_ratings_df['Sum']==0]
```

```python
#Ignoring the users whose total rating is zero
each_user_sum_of_ratings_df = each_user_sum_of_ratings_df[each_user_sum_of_ratings_df['Sum'] != 0]
```

```python
each_user_sum_of_ratings_df.shape
```

```python
each_user_sum_of_ratings_df.head()
```

```python
# finally on every vector genre rating / total sum of ratings ( normalization)
each_user_vector_genre_rating_df = each_user_sum_of_ratings_df.copy()
each_user_vector_genre_rating_df[["Horror","Romance","Thriller","comics","Graphics","Adventure","Fantasy","Mystery","Crime"]] = each_user_sum_of_ratings_df.loc[:,"Horror":"Crime"].div(each_user_sum_of_ratings_df['Sum'],axis=0)
each_user_vector_genre_rating_df
```

```python
each_user_vector_genre_rating_df.shape
```

```python
#fetch the highest weighed genre for every user and created user : genre list
each_user_most_liked_genre_df = each_user_vector_genre_rating_df.copy()
each_user_most_liked_genre_df['liked_genre'] = each_user_vector_genre_rating_df[["Horror","Romance","Thriller","comics","Graphics","Adventure","Fantasy","Mystery","Crime"]].idxmax(axis=1)
```

```python
#Here we can infer thatall the users have highest interest on Horror. This is because the dataset for genre Horror is huge compeared to other genres.
each_user_most_liked_genre_df[each_user_most_liked_genre_df.liked_genre !='Horror']
```

```python
each_user_vector_genre_rating_df.columns
```

```python
df_userprofile = each_user_vector_genre_rating_df[['user_id', 'Horror', 'Romance', 'Thriller', 'comics', 
                                                   'Graphics', 'Adventure', 'Fantasy', 'Mystery', 'Crime']]
```

```python
df_userprofile.shape
```
