# Cross Domain Recommender Systems

In this project we are perfoming recommendations across different domains. The domains we considers for the project are book reading and movies.
As initial step , this notebook has the details of preparing data for the genres Romance , Horror_Fantasy, Thriller

```python
import pandas as pd
import json
```

uploading the data files:  2 files for each genre
Romance: goodreads_books_romance.json (for book_id details) , goodreads_reviews_romance.json (for userID details)
Horror: goodreads_books_fantasy_paranormal.json (for book_id details), goodreads_reviews_fantasy_paranormal.json(for userID details)
Thriller: goodreads_books_mystery_thriller_crime.json(for book_id details), 'goodreads_reviews_mystery_thriller_crime.json'(for userID details)

goodreads dataset reference : https://sites.google.com/eng.ucsd.edu/ucsdbookgraph/home

```python
df = pd.read_json('goodreads_books_romance.json', lines=True)
df2 = pd.read_json('goodreads_reviews_romance.json', lines=True)
df3 = pd.read_json('goodreads_books_mystery_thriller_crime.json', lines=True)
df4 = pd.read_json('goodreads_reviews_mystery_thriller_crime.json', lines=True)
df5 = pd.read_json('goodreads_books_fantasy_paranormal.json', lines=True)
df6 = pd.read_json('goodreads_reviews_fantasy_paranormal.json', lines=True)
```

# Preparing dataset for Genre : Romance 

```python
#Displaying first five rows of romance dataset
df.head()
```

```python
#Displaying original columns
df.columns
```

```python
#Taking only the columns required for achieving our goal
df = df[['book_id', 'average_rating', 'description','title', 'language_code']]
```

```python
#Inserting genre column in the dataframe, this dataset is romance
df['Genre'] = 'Romance'
```

```python
df.head()
```

```python
#Displaying the laguage code values available in the dataset
df['language_code'].unique()
```

```python
#Removing all other laguage books except English
df_en = df[df['language_code'].isin(['en-US','eng','en-GB','en-CA','en'])]
```

```python
df_en['language_code'].unique ()
```

```python
df_en.shape
```

```python
df_en.head()
```

```python
#getting the size of romance reviews dataset 
df2.shape
```

```python
#displaying the columns
df2.columns
```

```python
df2.head()
```

```python
#checking one specific book_id value in romance reviews dataset. we can observe here one book_id has multiple reviews
df2.loc[df2['book_id'] == 29074693]
```

```python
#Merging the Romance dataset and romance review dataset with 
#respect to book_id and retrieving user_id, rating, review_text from romance reviews dataset
romance = pd.merge(df_en, df2[['user_id', 'book_id', 'rating', 'review_text']], on='book_id')
```

```python
romance.shape
```

```python
romance.head()
```

```python
#corss checking with previous book_id reference on the final romance dataset after merging
romance.loc[romance['book_id'] == 29074693]
```

```python
romance.columns
```

```python
romance.tail()
```

```python
#Converting the file to csv format for further computations
romance.to_csv('romance_final.csv')  # --> File number 1.
```
